```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('maoyan_2.csv',encoding='utf-8')
```

```python
data.head()
```

```python
data['long_time'] = data['long_time'].str.strip('分钟')
```

```python
data.head()
```

```python
data.info()
```

```python
data['time'] = data['time'].str.strip('上映时间：')
data
```

```python
data.head()
```

```python
data['year'] = data['time'].str.split('-').str[0]
data['month'] = data['time'].str.split('-').str[1]
data.head()
```

```python
data['star'] = data['star'].str.strip('主演：')
data.head()
```

```python
data.info()
```

```python
data['long_time'] = data['long_time'].astype('int64')
data.dtypes
```

```python
data.head()
```

```python
def get_country(i):
    name = i.split('(')
    if len(name) == 1:
        return '中国'
    else:
        county_1 =  name[1].strip(')')
        if county_1 == '中国香港':
            return '中国'
        elif county_1 == '法国戛纳':
            return '法国'
        else:
            return county_1
```

```python
data['country'] = data['time'].map(get_country)
data
```

```python
data['country'].unique()
```

```python
data.head()
```

```python
year_count = data.groupby('year')['year'].count()
year_count
```

```python
from pyecharts import Line

attr = list(year_count.index)
v1 = list(year_count)
line = Line("电影时间趋势")
line.add("电影", attr, v1, is_smooth=True, mark_line=["max", "average"])
line
```

```python
str = ''
for i in range(100):
    str = str + ',' + data.iloc[i,4]
str
```

```python
style_1 = str.split(',')
style = style_1[1:]
style
```

```python
style.count('爱情')
```

```python
from collections import Counter
c = Counter(style)
c
```

```python
style_count = c.most_common(6)
style_count
```

```python
v = []
attr = []
for i in style_count:
    v.append(i[1])
    attr.append(i[0])
print(v,attr)
```

```python
from pyecharts import Pie
pie = Pie("电影类型分布", title_pos='center', width=900)
pie.add(
    "",
    attr,
    v,
    is_random=True,
    radius=[30, 75],
    rosetype="area",
    is_legend_show=False,
    is_label_show=True,
)
pie
```
